```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
df = pd.read_csv("./data/Bengaluru_House_Data.csv")
# df = df.head(50)
```

```python
df.isnull().sum()
```

```python
df['price'].describe()
```

```python
# Identify Rows with Non-Numeric Units
# Use regex to find rows with alphabetic characters (units)
# Keep rows with only digits, commas, or dashes (no letters)
df = df[~df['total_sqft'].str.contains('[a-zA-Z]', na=False, regex=True)]
```

```python
# Function to convert ranges to their average value
def clean_sqft(x):
    if isinstance(x, str):
        if '-' in x:  # Handle ranges (e.g., "1000-1500")
            low, high = map(float, x.split('-'))
            return (low + high) / 2
        elif ',' in x:  # Handle commas (e.g., "1,200")
            return float(x.replace(',', ''))
    return float(x)

# Apply the conversion
df['total_sqft'] = df['total_sqft'].apply(clean_sqft)
```

```python
# Remove flat with 1 sq ft data
df = df[df['total_sqft'] > 300]
```

```python
# Conversion of Size to numeric(BHK)
def string_to_numeric_converter(size):
    if pd.isna(size):
        return np.nan
    bhk = size.split(' ')[0] if isinstance(size, str) else str(size)
    return int(bhk)
```

```python
df['size'] = df['size'].apply(string_to_numeric_converter)
```

```python
# Drop unnecessary columns and rows with missing values
df = df.drop(['society','balcony','area_type', 'availability', 'location'], axis='columns')
```

```python
# Handle missing values (if any)
df = df.dropna()
```

```python
df.head(2)
```

```python
# Data Exploration: Before modeling, always check your data:
# print(df['total_sqft'].unique())  # See all unique values
# print(df['total_sqft'].describe())  # Summary statistics
```

```python
# Visualize After Cleaning:

# plot a scatter plot
# Set the figure size (e.g., 10 inches wide, 8 inches tall)
# Create a figure with a specified size (e.g., 10 inches wide, 6 inches tall)

plt.figure(figsize=(15,7))
plt.scatter(df.total_sqft, df.price, color="red", marker="+")
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (USD)')
plt.title('House Prices vs Area')
plt.show()
```

```python
# Feature Engineering - helps in removing outliers
df['price_per_sqft'] = df['price']*100000 / df['total_sqft']  # Convert price to rupees and calculate per sqft
```

```python
df.head(2)
```

```python
# Define features (X) and target (Y)
X = df[['size', 'total_sqft', 'bath']]
Y = df['price']
```

```python
# Train test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Scale features
# scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_test_scaled = scaler.transform(X_test)
```

```python
# Train model
reg = linear_model.LinearRegression()
reg.fit(X_train, Y_train)
```

```python
# Make predictions
Y_pred = reg.predict(X_test)
```

```python
# Evaluate

mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
# Predict a new house price
def predict_price(size, total_sqft, bath):
    features = np.array([[size, total_sqft, bath]])
    predicted_price = reg.predict(features)
    return predicted_price[0]
```

```python
# Example prediction - size, total_sqft, bath
sample_pred = predict_price(4, 2600.0, 5)
print(f"Predicted Price for 4 BHK, 2600.0 sqft, 5 baths: ₹{sample_pred:.2f} lakhs")
```

```python
# Visualize predictions vs actual (for total_sqft)
plt.figure(figsize=(10, 6))
plt.scatter(X_test['total_sqft'], Y_test, color='blue', label='Actual Price')
plt.scatter(X_test['total_sqft'], Y_pred, color='red', label='Predicted Price')
plt.xlabel('Total Sqft')
plt.ylabel('Price (in lakhs)')
plt.title('Actual vs Predicted House Prices')
plt.legend()
plt.show()
```
