```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# Add model evaluation
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("./data/canada_per_capita_income.csv")
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={'per capita income (US$)': 'income'})
```

```python
df.head(2)
```

```python
# Define features (X) and target (Y)
X = df[['year']]  # Note the double brackets to keep it as a DataFrame
Y = df['income']
```

```python
# Train test split
# X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Train model
model = LinearRegression()
model.fit(X, Y)
```

```python
# Check the model coefficients
print(f"Intercept (baseline income): {model.intercept_:.2f}")
print(f"Slope (income increase per year): {model.coef_[0]:.2f}")
```

```python
# 4. Test predictions on existing data
y_pred = model.predict(X)  # Predict for all years in dataset
df['predicted'] = y_pred   # Add predictions to DataFrame
df.head() # Compare actual vs predicted
```

```python
# 5. Evaluate model performance
r2 = r2_score(Y, y_pred)
print(f"\nR² Score (0-1, higher is better): {r2:.4f}")
```

```python
# 6. Predict for future years (2020, 2021, 2022, 2023, 2024)
future_years = [[2020], [2021], [2022], [2023], [2024]]
future_predictions = model.predict(future_years)
print(f"\nPredicted Income (2020): ${future_predictions[0]:.2f}")
print(f"Predicted Income (2021): ${future_predictions[1]:.2f}")
print(f"\nPredicted Income (2022): ${future_predictions[2]:.2f}")
print(f"Predicted Income (2023): ${future_predictions[3]:.2f}")
print(f"\nPredicted Income (2024): ${future_predictions[4]:.2f}")
print(f"Predicted Income (2025): ${future_predictions[5]:.2f}")
```
